```python
import numpy as np
from scipy.stats import wilcoxon
import matplotlib.pyplot as plt

N = 10

data1 = np.random.normal(loc = 0.0, scale = 1.0, size = N)
data2 = np.random.normal(loc = 0.0, scale = 1.0, size = N)

plt.scatter(data1, np.zeros(N), c = 'r') # plot first line
plt.scatter(data2, np.zeros(N) + 1, c = 'b') # plot first line


datax = [data1, data2]
datay = [np.zeros(N), (np.zeros(N) + 1)]
_ = plt.plot(datax, datay, 'yo', linestyle="--", alpha = 0.2)

stat, p = wilcoxon(data1, data2, mode="approx")

print('p-value=%.3f' % (p))

if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```

```python
N = 20

for i in range(100):
    data1 = np.random.normal(loc = 0.0, scale = 1.0, size = N)
    data2 = np.random.normal(loc = 0.0, scale = 1.0, size = N)
    
    stat, p = wilcoxon(data1, data2, mode="approx")
    
    if p > 0.05:
        #print('Probably the same distribution')
        continue
    else:
        print('Probably different distributions ' + str(i))
        print('p-value = ' + str(p))
        #print(stat, p)
        continue

plt.scatter(data1, np.zeros(N), c = 'r') # plot first line
plt.scatter(data2, np.zeros(N) + 1, c = 'b') # plot first line


datax = [data1, data2] #, np.concatenate(np.zeros[10], np.zeros[10] + 1)] 
datay = [np.zeros(N), (np.zeros(N) + 1)]
_ = plt.plot(datax, datay, 'yo', linestyle="--", alpha = 0.2)
```
