```python
import numpy as np

np.random.seed(123)

x = 5 * np.random.rand(10)
y = 2 * np.random.rand(10)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, 
         color = 'red', 
         linewidth = 0.1, 
         linestyle = '--', 
         marker = '.')
```

```python
plt.plot(x, y, color = 'red', linewidth = 0.5, linestyle = ':', marker = '.')
plt.scatter(x, y, color = 'yellow')
```

```python
from scipy.stats import norm

np.random.seed(123)

# generate a random distribution with mean 1.0 and stddev 0.2
x = np.random.normal(loc = 1.0, scale = 0.2, size = 100)
# generates the probability density function of the distribution x
y = norm.pdf(x, loc = 1.0, scale = 0.2)

plt.hist(x, density = True, alpha = 0.5)
plt.scatter(x, y, color = "red")
```

```python
y_target = 2 * x + 5

plt.scatter(x, y_target, color = "green")
```

```python
from scipy.stats import expon

x = np.random.exponential(scale = 0.2, size = 100)

y = expon.pdf(x, scale = 0.2)

plt.hist(x, density = True, alpha = 0.5)
plt.scatter(x, y, color = "red")

```

```python
y_target = 2 * x + 5

plt.scatter(x, y_target, color = "green")
```
