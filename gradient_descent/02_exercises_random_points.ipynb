```python
import numpy as np
import matplotlib.pyplot as plt

# set random seed
np.random.seed(123)

# completely random points
x = np.random.rand(50)
y = np.random.rand(50)

plt.plot(x, y, color = 'red', marker = '.', linestyle = ':', linewidth = 0.01)
```

```python
# set random seed
np.random.seed(123)

# gaussian distribution of points in a line
x = np.random.rand(100)

# target function UNKNOWN!!!!
y = 2 * x

plt.plot(x, y, color = 'red', marker = '.', linestyle = ':', linewidth = 0.01)
```

```python
np.random.seed(123)

N = 100

x = np.random.normal(loc=0.5, scale=1.0, size=N)

noise = np.random.normal(loc = 0, scale=0.1, size=N)

# target function UNKNOWN!!!!
y = 5 * x + noise

plt.plot(x, y, color = 'red', marker = '.', linestyle = ':', linewidth = 0.01)
```
